```python
import sklearn.datasets as datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
customer=np.random.randn(1,1000)
```

```python
sns.distplot(customer)
```

```python
purchases=customer*13+9
```

```python
sns.distplot(purchases)
```

```python
#we have two  variables dependent on each other linearly
```

```python
dfx=pd.DataFrame(customer.T)
```

```python
dfy=pd.DataFrame(purchases.T)
```

```python
#now creating linear regression model using sklearn
```

```python
model=LinearRegression()
```

```python
model.fit(dfx,dfy)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
ypred=model.predict(dfx)
```
