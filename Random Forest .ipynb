```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

```python
data=load_iris()
```

```python
data.feature_names
```

```python
data.target_names
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
model=RandomForestClassifier(n_estimators=100,n_jobs=2)
```

```python
model.fit(df,data.target)
```

```python
y=model.predict(df)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data.target,y)
```
